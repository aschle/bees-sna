```python
def get_distances(df):
    df = df.reset_index(level = 'frame_idx')
    m = pd.merge(df, df, on='frame_idx')
    m = m[m.id_x < m.id_y]
    m.loc[:, 'dist'] = np.sqrt(np.square(m.xpos_x - m.xpos_y) \
        + np.square(m.ypos_x - m.ypos_y))
    return m
```
