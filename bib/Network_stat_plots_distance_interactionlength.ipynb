```python
import numpy as np
import pandas as pd

from pandas import DataFrame
import seaborn as sns
import ast

%matplotlib inline
import matplotlib.pyplot as plt
```

# Distance (radius)

```python
df = pd.read_csv("1h_09_95_3L_distance.csv", sep=";")
```

```python
#df.plot.scatter(x='dist', title="T", subplots=True, layout=(1,2))

fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(10,18), )

df.plot.scatter(ax=axes[0,0], x='dist', y='nodes')
axes[0,0].set_title('#Nodes')

df.plot.scatter(ax=axes[0,1], x='dist', y='edges')
axes[0,1].set_title('#Edges')

df.plot.scatter(ax=axes[1,0], x='dist', y='av_deg')
axes[1,0].set_title('Average Degree')

df.plot.scatter(ax=axes[1,1], x='dist', y='density')
axes[1,1].set_title('Desity')

df.plot.scatter(ax=axes[2,0], x='dist', y='av_shortest_path')
axes[2,0].set_title('Average Shortes Path')

df.plot.scatter(ax=axes[2,1], x='dist', y='cc')
axes[2,1].set_title('Clustering Coefficient')

df.plot.scatter(ax=axes[3,0], x='dist', y='components')
axes[3,0].set_title('#Components')

df.plot.scatter(ax=axes[3,1], x='dist', y='diameter')
axes[3,1].set_title('Diameter')
```

# Distance - Degree Distribution

```python
df.degree = df.degree.apply(lambda x: ast.literal_eval(x.split("(")[1].split(")")[0]))
```

```python
fig, axes = plt.subplots(5,4)

for i, r in df.iterrows():
    nrow = int(i/4)
    ncol = i%4
    axes[nrow,ncol].hist(r.degree, bins=20)
    axes[nrow,ncol].set_title("dist={}".format(r.dist))
    axes[nrow,ncol].set_xlim(0,280)
    axes[nrow,ncol].set_ylim(0,110)

fig.set_size_inches(30,30)
plt.show()
```
